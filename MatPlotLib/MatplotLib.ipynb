```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-20, 20, 1)
w1 = np.array([3, 4, -10])
p = np.array([4, 5, 1])

print ('x: {}'.format(x.shape))
print ('w1: {}'.format(w1.shape))
print ('p: {}'.format(p.shape))
```

```python
def f(x, w):
    return -(w[0]/w[1])* x -(w[2]/w[1])
```

```python
print (f(x, w1))
```

```python
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(20)

ax.plot(x, f(x, w1), 'r', label="Before", linewidth=10.0)
for i in range(1, 10):
    a = i/5
    w = w1 - a*p
    ax.plot(x, f(x, w), label="Learning Rate = {}".format(a))
    
ax.plot([4], [5], '^', label="Point")
ax.set_xlim(-15, 15)
ax.set_ylim(-15, 15)
ax.legend()
ax.grid(True)
        



'''plt.plot(x1, f1(x1), 'b', x1, f2(x1), 'r')
plt.scatter([4], [5])
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.grid(True)
plt.xticks()'''
```
